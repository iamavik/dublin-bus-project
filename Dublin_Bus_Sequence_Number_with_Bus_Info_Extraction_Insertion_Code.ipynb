```python
result = testing_with_stops_times_csv()
```

```python
i=0
new_data_list = [0]*len(result)
for data in result:
    trip_id = data[0]
    x = trip_id.find("-")
    y = trip_id.find("-",x+1)
    bus_number= trip_id[x+1:y]
    new_data_list[i] = list(result[i])
    new_data_list[i].append(bus_number)      
    i+=1
    
```

```python
used = set()
unique = [x for x in new_data_list if x[1]+str(x[2])+x[6] not in used and ((used.add(x[1]+str(x[2])+x[6])) or True)]

```
